```
#Matplotlib

import numpy as np
import matplotlib.pyplot as plt
```

```
a= np.arange(0,3*np.pi,0.5)
b=np.sin(a)
plt.plot(a,b)
plt.show()
```

```
data= np.arange(0,3*np.pi,0.5)
sin_data=np.sin(data)
cos_data= np.cos(data)

plt.subplot(1,2,1)
plt.plot(data,sin_data)
plt.title('Sine')
plt.plot(1,2,2)
plt.plot(data,cos_data)
plt.title('Cosine')
plt.xlabel(['X axis data'])
plt.ylabel('Y axis data')
plt.legend(['Sine','Cosine'])
plt.title('Sine VS Cosine')
plt.show()

```
